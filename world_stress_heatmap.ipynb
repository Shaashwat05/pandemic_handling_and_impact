```python
import pandas as pd
from json import load
```

```python
df = pd.read_csv("COVIDiSTRESS_April_May_Combined.csv")
df.head()
```

```python
from geopy.geocoders import Nominatim

geolocator = Nominatim(user_agent="geoapiExercises")
country = []
for row in df.iterrows():
    location = geolocator.reverse(str(row[1][2]*90.0)+","+str(row[1][3]*180.0))
    country.append(location.raw['address'].get('country', ''))
country
```

```python
df['country'] = country
df.head()
```

```python
df_country = df.groupby(['country']).mean()
```
